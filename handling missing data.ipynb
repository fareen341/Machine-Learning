# Imputing the missing values:

1)with 0.

2)with some name.

3)with mean.

4)with median.

5)with Forward fill(usefull in time series).

6)with Backward fill(usefull in time series).

7)Replacing with average of previous and next value(usefull in time series).


```python
#replacing with 0 
df=pd.read_csv("some data.csv")
df['Some column']= df['Some column'].fillna(0)
df
```

```python
#replacing with lets say 'mumbai'
df=pd.read_csv("some data.csv")
df['Some column']= df['Some column'].fillna('Mumbai')
df
```

```python
#replacing with mean
df=pd.read_csv("some data.csv")
df['Height']=df['Height'].fillna((df['Height'].mean()))
df
```

```python
#replacing with median
df=pd.read_csv("some data.csv")
df['Weight']=df['Weight'].fillna((df['Weight'].median()))
df
```

```python
#Replacing with previous value - Forward fill
#In time series data, replacing with nearby values will be more appropriate than replacing it with mean. 
#Forward fill method fills the missing value with the previous value.
df=pd.read_csv("some data.csv")
df['Number of days'] = df['Number of days'].fillna(method='ffill')
df
```

```python
#Replacing with next value - Backward fill
#Backward fill uses the next value to fill the missing value.
#usefull in timeseries.
df=pd.read_csv("some data.csv")
df['Number of days'] = df['Number of days'].fillna(method='bfill')
df
```

```python
#Replacing with average of previous and next value
#In time series data, often the average of value of previous and next value will be a better estimate of the missing value.
df=pd.read_csv("some data.csv")
df['Number of days'] = pd.concat([df['Number of days'].ffill()
                                       , df['Number of days'].bfill()]).groupby(level=0).mean()
df
```

# Removing the missing values:
MCAR: Missing completely at random, this is great. It means that the non-response is completely random. 
    So your survey is not biased.

MAR: Missing at random, worse situation. Imagine you are asking for IQ and you have much more females participants than males. 
    Lucky for you, IQ is not related to gender, so you can control for gender (apply weighting) to reduce bias.

MNAR: Not missing at random, bad. Consider having survey for level of income. 
    And again, you have more females than males participants. In this case, this is a problem, because level of income is related to gender. 
    Therefore your results will be biased. Not easily to get rid of.
    
1)Dropping the rows having null values.

2)Dropping the entire column having null values.

```python
#dropping the row
df=pd.read_csv("some data.csv")
miss_data=df.dropna()
miss_data
```

```python
#dropping the entire column
df=pd.read_csv("some data.csv")
cdata= df.drop(['Some column'], axis=1)
cdata
```
